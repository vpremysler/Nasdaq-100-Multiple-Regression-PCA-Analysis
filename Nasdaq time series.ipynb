```python

!pip install pytrends
!pip install pandas
!pip install requests
!pip install fredapi
!pip install statsmodels
!pip install scikit-learn

# Import libraries
import yfinance as yf
import pandas as pd
import pandas_datareader as pdr
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import os
import requests
from fredapi import Fred
from pytrends.request import TrendReq
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Set time range
start_date = '2010-01-01'
end_date = datetime.datetime.now().strftime('%Y-%m-%d')

# FRED API key
fred_api_key = '13b3e8ef4a8b58c3015e9e7c523af730'
fred = Fred(api_key=fred_api_key)

# Consumer Sentiment Index data
consumer_sentiment = fred.get_series('UMCSENT', observation_start='2010-01-01')
consumer_sentiment_df = pd.DataFrame(consumer_sentiment, columns=['Consumer Sentiment'])

# Construction Spending data
construction_data = fred.get_series('TTLCONS', observation_start='2010-01-01')
construction_data_df = pd.DataFrame(construction_data, columns=['Construction'])

# Set up pytrends for Google Trends data
pytrends = TrendReq(hl='en-US', tz=360)

# Define the search terms
keywords = ["vacation", "new job", "recession"]
pytrends.build_payload(keywords, timeframe='2010-01-01 2024-06-17')
trends_data = pytrends.interest_over_time().drop(columns=['isPartial'])

# Combine all data into a single DataFrame
combined_df = consumer_sentiment_df.join(construction_data_df, how='outer')
combined_df = combined_df.join(trends_data, how='outer')

# Define tickers
tickers = {
    'QQQ': 'QQQ',
}

# Download market data
market_data = {name: yf.download(ticker, start=start_date, end=end_date) for name, ticker in tickers.items()}

# Define macroeconomic data sources and symbols
macro_symbols = {
    'CPI': 'CPIAUCSL',
    'Interest Rate': 'DFF',
    'Unemployment': 'UNRATE',
}

# Download economic data
econ_data = {name: pdr.get_data_fred(symbol, start=start_date, end=end_date) for name, symbol in macro_symbols.items()}

# Add Google Trends data
econ_data['Google Trends: Vacation'] = trends_data['vacation']
econ_data['Google Trends: New Job'] = trends_data['new job']
econ_data['Google Trends: Recession'] = trends_data['recession']

# Add Consumer Sentiment and Construction Data
econ_data['Consumer Sentiment'] = consumer_sentiment_df
econ_data['Construction'] = construction_data_df

# Create directories to save data
if not os.path.exists('market_data'):
    os.makedirs('market_data')
if not os.path.exists('econ_data'):
    os.makedirs('econ_data')

# Save data to csv files
for name, df in market_data.items():
    df.to_csv(f'market_data/{name}.csv')
for name, df in econ_data.items():
    df.to_csv(f'econ_data/{name}.csv')

# Resample QQQ to get the last closing price of each month using the last business day
market_close_monthly = market_data['QQQ']['Close'].resample('BM').last()

# Align the data
econ_data_combined = pd.concat(econ_data.values(), axis=1)
econ_data_combined.columns = econ_data.keys()
econ_data_monthly = econ_data_combined.resample('M').last()

# Combine data
combined_data = market_close_monthly.to_frame().join(econ_data_monthly, how='inner')

# Handle missing values
combined_data = combined_data.ffill().dropna()

# Plot the data on separate subplots
fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(14, 24))

# Plot QQQ
axes[0].plot(combined_data.index, combined_data['Close'], label='QQQ', color='blue')
axes[0].set_title('QQQ Stock Price Over Time')
axes[0].legend()

# Plot Consumer Sentiment
axes[1].plot(combined_data.index, combined_data['Consumer Sentiment'], label='Consumer Sentiment', color='blue')
axes[1].set_title('Consumer Sentiment Index')
axes[1].legend()

# Plot Construction Spending
axes[2].plot(combined_data.index, combined_data['Construction'], label='Construction Spending', color='orange')
axes[2].set_title('Construction Spending')
axes[2].legend()

# Plot Vacation Search
axes[3].plot(combined_data.index, combined_data['Google Trends: Vacation'], label='Vacation Search', color='green')
axes[3].set_title('Vacation Search Trends')
axes[3].legend()

# Plot New Job Search
axes[4].plot(combined_data.index, combined_data['Google Trends: New Job'], label='Google Trends: New Job', color='red')
axes[4].set_title('New Job Search Trends')
axes[4].legend()

# Plot Recession Search
axes[5].plot(combined_data.index, combined_data['Google Trends: Recession'], label='Google Trends: Recession', color='purple')
axes[5].set_title('Recession Search Trends')
axes[5].legend()


plt.tight_layout()
plt.show()

# Calculate and plot the correlation matrix
correlation_matrix = combined_data.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Calculate VIF for independent variables
def calculate_vif(df):
    vif = pd.DataFrame()
    vif["Variable"] = df.columns
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif

# Initial independent variables
independent_vars = combined_data[['Consumer Sentiment', 'Construction', 'Google Trends: Vacation', 'Google Trends: New Job', 'Google Trends: Recession']]

# Perform multiple regression analysis
def perform_multiple_regression(independent_vars, dependent_var, labels):
    independent_vars = sm.add_constant(independent_vars)  # Adds a constant term to the independent variables
    model = sm.OLS(dependent_var, independent_vars).fit()  # Fit the model
    print(f"Multiple Regression Results for {labels[-1]} vs {', '.join(labels[:-1])}")
    print(model.summary())  # Print the summary of the regression

    # Plot the regression results
    plt.figure(figsize=(14, 7))
    plt.scatter(model.fittedvalues, dependent_var, label='Observed', color='blue', alpha=0.5)
    plt.plot(model.fittedvalues, model.fittedvalues, label='Fitted', color='red')
    plt.xlabel('Fitted Values')
    plt.ylabel(labels[-1])
    plt.title(f'Multiple Regression: {labels[-1]} vs {", ".join(labels[:-1])}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Align the data for regression
qqq = combined_data['Close']

# Perform multiple regression with all five variables
perform_multiple_regression(independent_vars, qqq, list(independent_vars.columns) + ['QQQ'])

# Calculate initial VIF
vif = calculate_vif(independent_vars)
print("Initial VIF:")
print(vif)


# Standardize the independent variables
scaler = StandardScaler()
independent_vars_scaled = pd.DataFrame(scaler.fit_transform(independent_vars), index=independent_vars.index, columns=independent_vars.columns)

# Perform PCA
pca = PCA(n_components=3)  # Adjust the number of components as needed
principal_components = pca.fit_transform(independent_vars_scaled)
principal_df = pd.DataFrame(data=principal_components, index=independent_vars.index, columns=['PC1', 'PC2', 'PC3'])

# Perform multiple regression with principal components
perform_multiple_regression(principal_df, qqq, list(principal_df.columns) + ['QQQ'])






# remove variables and perform regression
def iterative_variable_removal(independent_vars, dependent_var):
    variables = independent_vars.columns.tolist()
    while len(variables) > 1:
        print(f"\nRegression with variables: {variables}")
        perform_multiple_regression(independent_vars[variables], dependent_var, variables + ['QQQ'])
        vif = calculate_vif(independent_vars[variables])
        print("VIF:")
        print(vif)

        # Identify the variable with the highest VIF
        max_vif_var = vif.loc[vif['VIF'].idxmax(), 'Variable']

        # Check if the highest VIF is above an acceptable threshold
        if vif['VIF'].max() < 8:
            break

        # Remove the variable with the highest VIF
        print(f"Removing variable with highest VIF: {max_vif_var}\n")
        variables.remove(max_vif_var)

    if len(variables) > 0:
        print("\nFinal Model:")
        perform_multiple_regression(independent_vars[variables], dependent_var, variables + ['QQQ'])
        vif = calculate_vif(independent_vars[variables])
        print("Final VIF:")
        print(vif)
    else:
        print("No variables left for regression.")

# Perform iterative variable removal
iterative_variable_removal(independent_vars, qqq)






# perform simple linear regression and plot results
def simple_regression(independent_var, dependent_var, independent_var_name):
    independent_var = sm.add_constant(independent_var)  # Adds a constant term to the independent variable
    model = sm.OLS(dependent_var, independent_var).fit()  # Fit the model
    print(f"Simple Regression Results for QQQ vs {independent_var_name}")
    print(model.summary())  # Print the summary of the regression

    # Plot the regression results
    plt.figure(figsize=(14, 7))
    plt.scatter(independent_var[independent_var_name], dependent_var, label='Observed', color='blue', alpha=0.5)
    plt.plot(independent_var[independent_var_name], model.fittedvalues, label='Fitted', color='red')
    plt.xlabel(independent_var_name)
    plt.ylabel('QQQ Closing Price')
    plt.title(f'Simple Regression: QQQ vs {independent_var_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Independent variables
independent_vars = combined_data[['Consumer Sentiment', 'Construction', 'Google Trends: Vacation', 'Google Trends: New Job', 'Google Trends: Recession']]

# Align the data for regression
qqq = combined_data['Close']

# Perform simple regression for each independent variable
for var in independent_vars.columns:
    simple_regression(independent_vars[[var]], qqq, var)

```
